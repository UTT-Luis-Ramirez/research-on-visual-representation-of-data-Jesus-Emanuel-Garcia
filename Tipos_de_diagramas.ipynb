# **HISTOGRAMA**

###¿Cómo se crea?

Agrupar los datos: Se dividen los datos en intervalos o rangos.
Contar las frecuencias: Se cuenta cuántos datos caen dentro de cada intervalo.
Crear el gráfico: Se dibujan barras donde la base representa el intervalo y la altura, la frecuencia.

###¿Para qué sirve?

Visualizar la distribución de los datos: Permite ver rápidamente si los datos están distribuidos de manera uniforme, si hay valores atípicos, etc.
Identificar patrones: Ayuda a detectar tendencias o patrones en los datos.
Comparar conjuntos de datos: Se pueden comparar dos o más histogramas para ver las diferencias entre ellos.

###¿En qué contexto se utiliza?

Los histogramas son muy útiles en diversos campos, como:

**Estadística**: Para analizar y describir conjuntos de datos.

**Ciencia de datos**: Para explorar y visualizar datos antes de realizar análisis más complejos.

**Control de calidad**: Para monitorear procesos y detectar variaciones.

**Investigación**: Para presentar resultados de manera visual y comprensible.

# DIAGRAMA DE TALLO Y HOJA

###**¿Cómo se crea?**

Organiza los datos: Ordena los datos de menor a mayor.
Separa los dígitos: Separa cada dato en tallo y hoja.
Construye el diagrama: Escribe los tallos en una columna y las hojas correspondientes a cada tallo a su lado.


###**¿Para qué sirve?**

**Visualizar la distribución de los datos**: Permite ver rápidamente si los datos están concentrados en un rango específico, si hay valores atípicos, etc.

**Ordenar los datos**: Al construir el diagrama, los datos quedan ordenados automáticamente.

**Calcular medidas descriptivas**: A partir del diagrama se pueden calcular fácilmente la mediana, la moda y el rango.

**Comparar conjuntos de datos**: Se pueden construir dos diagramas lado a lado para comparar la distribución de dos conjuntos de datos.

###**¿En qué contexto se utiliza?**

**Estadística descriptiva**: Para representar y analizar datos de forma sencilla.

**Educación**: Para enseñar conceptos básicos de estadística.

**Investigación**: Como una primera aproximación al análisis de datos.

Imaginemos que tenemos las siguientes edades: 15, 18, 22, 22, 25, 28, 30.

#DIAGRAMA DE CAJA

###¿Cómo se crea?

**Calcular los cuartiles**: Se ordenan los datos de menor a mayor y se calculan el primer cuartil (Q1), la mediana (Q2) y el tercer cuartil (Q3).

**Determinar el rango intercuartílico (IQR)**: Se calcula restando Q1 a Q3.

**Identificar los valores atípicos**: Generalmente, se consideran valores atípicos aquellos que están a más de 1.5 veces el IQR por encima de Q3 o por debajo de Q1.

**Dibujar la caja**: Se dibuja un rectángulo donde los lados izquierdo y derecho corresponden a Q1 y Q3, respectivamente, y una línea vertical dentro de la caja representa la mediana (Q2).

**Dibujar los bigotes**: Se trazan líneas desde los extremos de la caja hasta el valor mínimo y máximo que no sean considerados como valores atípicos. Los valores atípicos se representan con puntos individuales.

###**¿Para qué sirve?**

**Visualizar la distribución de los datos**: Permite ver rápidamente la forma de la distribución, si es simétrica o asimétrica, y si hay valores atípicos.

**Comparar diferentes conjuntos de datos**: Se pueden comparar varios diagramas de caja para identificar diferencias en las distribuciones.

**Identificar valores atípicos**: Los valores atípicos se destacan visualmente y pueden ser objeto de análisis posterior.

**Resumen de los datos**: Proporciona un resumen visual de cinco números clave: mínimo, primer cuartil, mediana, tercer cuartil y máximo.

###**¿En qué contexto se utiliza?**

**Estadística descriptiva**: Para explorar y visualizar datos.

**Análisis de datos**: Para identificar patrones, tendencias y valores atípicos.

**Control de calidad**: Para monitorear procesos y detectar variaciones.

**Investigación**: Para presentar resultados de manera visual y comprensible.
